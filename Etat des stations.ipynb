# Visualisation des stations cy'clic de Rouen

```python
import pandas as pd
import folium
import requests
import json
```

```python
rouen_coordinates = (49.4385,1.0953)
m = folium.Map(location=rouen_coordinates,zoom_start=14)

r = requests.get('https://api.jcdecaux.com/vls/v3/stations?contract=rouen&apiKey=cd065b1b6a132d15cac706f9214f322f889e2741')
stations = []
for station in r.json():
    s = dict()
    s['number']=station['number']
    s['name']=station['name']
    s['address']=station['address']
    s['position']=station['position']
    s['status']=station['status']
    s['lastUpdate']=station['lastUpdate']
    s['connected']=station['connected']
    s['capacity']=station['totalStands']['capacity']
    s['bikes']=station['totalStands']['availabilities']['bikes']
    s['stands']=station['totalStands']['availabilities']['stands']
    
    location = [s['position']['latitude'],s['position']['longitude']]
    popup = s['address']+'\n Capacité : '+str(s['capacity'])+'\n Vélos disponibles : '+str(s['bikes'])+'\n Emplacements libres : '+str(s['stands'])
    folium.Marker(
        location=location,
        popup=popup,
        icon=folium.Icon()
    ).add_to(m)
    stations.append(s)

df = pd.DataFrame(stations)
m
```
